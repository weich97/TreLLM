Load the generated artifact and inspect the case-level metrics.

```python
import json
from pathlib import Path

path = Path("outputs/examples/notebook_trajectory.json")
payload = json.loads(path.read_text())
rows = []
for case_name, case_payload in payload.items():
    rows.append({"case": case_name, **case_payload["metrics"]})
rows
```
